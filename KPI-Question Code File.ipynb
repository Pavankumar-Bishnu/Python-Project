<code style="background:green;color:white">Question 12</code>

**How does exercise frequency relate to panic attack frequency or severity?**

```python
exercise_freq_corr = df["Exercise_Frequency"].corr(df["Panic_Attack_Frequency"])
exercise_severity_corr = df["Exercise_Frequency"].corr(df["Panic_Score"])
```

```python
print("Exercise >> Frequency correlation:" , round(exercise_freq_corr, 5))
print("Exercise >> Severity  correlation:" , round(exercise_severity_corr, 5))
```

<code style="background:green;color:white">Question 13</code>

**What percentage of smokers or heavy drinkers report severe panic attacks?**

```python
smokers_severe  = (df[(df["Smoking"] == "Yes") & 
                  (df["Panic_Score"] >= 8)].shape[0] / df[df["Smoking"] == "Yes"].shape[0]) * 100
drinkers_severe = (df[(df["Alcohol_Consumption"] >= 5) & 
                  (df["Panic_Score"] >= 8)].shape[0] /df[df["Alcohol_Consumption"] >5].shape[0]) * 100
```

```python
print("Smokers severe panic :", round(smokers_severe, 2), "%")
print("Heavy drinkers severe panic :", round(drinkers_severe, 2), "%")
```

<code style="background:green;color:white">Question 14</code>

**What is the age distribution of participants with high panic scores?**

```python
High_panic_age_mean   = df.loc[df["Panic_Score"] >= 8, "Age"].mean()
High_panic_age_median = df.loc[df["Panic_Score"] >= 8, "Age"].median()
High_panic_age_mode   = df.loc[df["Panic_Score"] >= 8, "Age_Category"].mode()
```

```python
print("High Panic Age mean:", round(High_panic_age_mean, 2))
print("High Panic Age median:", round(High_panic_age_median, 2))
print("High Panic Age mode:",High_panic_age_mode)
```
